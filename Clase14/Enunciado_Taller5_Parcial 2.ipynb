En el taller 4, iniciamos el preprocesamiento de los datos para poder realizar Sentiment Analysis.

En este entrega se debe:

1.   Terminar el preprocesamiento de los datos, el cual incluye:
*   Limpieza de carácteres básicos
*   Eliminar números
*   Unificar espacios en blanco
*   Eliminar puntuación
*   Eliminar los stopwords
En github hay un notebook `analisis_sentimientos_funciones` donde estan las funciones de interes para realizar este preprocesamiento.

2.   Realizar el Análisis de sentimientos decidiendo como se va a trabajar con los emojis, explorando 3 opciones:
*   Convertir los emojis a texto, puede hacerse utilizando la librería emoji,para procesarlos con el algoritmo de sentiment analysis
*   Eliminar los emojis del texto del mensaje y procesarlos con el algoritmo de sentiment analysis
*   Para los mensajes que tienen emojis, utilizar la libreria emosent solo con los emojis, calcular la polaridad de cada uno y promediarlos para establecer la polaridad del mensaje.

Lo anterior implica que se debe calcular tres scores para cada mensaje y tomando una muestra definir la utilidad de trabajar con los emojis para determinar la polaridad del mensaje.

3.  Se debe presentar un reporte (incluyendo gráficos) de cómo funcionan las 3 opciones de análisis de sentimientos y compararlas con el score original que incluía el dataset.
