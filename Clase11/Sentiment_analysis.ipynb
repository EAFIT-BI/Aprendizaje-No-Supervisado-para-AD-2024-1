```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Descarga el lexicon de vader
nltk.download('vader_lexicon')

```

```python
# Se carga el dataframe con los mensajes de texto
df = pd.read_csv('Elon_data.csv')

# Si los mensajes contienen caracteres diferentes a texto, se deben convertir a string
df['text'] = df['text'].astype(str)
```

```python
# Se inicializa vader
sid = SentimentIntensityAnalyzer()

# se crea una funci칩n para aplicar sentiment analysis y que entrega el compound score
def analyze_sentiment(message):
    scores = sid.polarity_scores(message)
    return scores['compound']

# Se aplica sentiment analysis a todos los mensajes de texto y se guarda el compound score en una nueva columna
df['sentiment_score'] = df['text'].apply(analyze_sentiment)

# Se guarda el dataframe que contiene la nueva columna en un nuevo archivo
df.to_csv('mensajes_procesados.csv', index=False)
```
