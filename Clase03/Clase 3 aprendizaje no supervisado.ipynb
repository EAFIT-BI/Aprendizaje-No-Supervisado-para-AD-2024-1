# Funciones de secuencia

En Python por defecto existe una función llamada range, que crea una secuencia entre dos núemros enteros, pero funciona como un intervalo \[inicio , fin\)

````python
range(5) # Secuencia de 0 hasta 4, 5 elementos
````

````python
list(range(5)) # Transformar la secuencia en una lista
````

````python
range(5,10) # Secuencia de números desde 5 hasta 9
````

````python
list(range(0,10) ) # Transformar la secuencia en una lista
````

También es posible realizar secuencias con numpy, donde se tienen mas variedad de opciones. También es de la misma manera de intervalo [inicio, fin), pero tiene la posibilidad de realizar un salto para la secuencia.

````python
np.arange(start= 1 ,stop=10 ) # Secuencia 1 hasta 9, se crea un vector.
````

````python
np.arange(start= 1 ,stop=50,step=5 ) # Secuencia 1 hasta 49, salto de 5.
````

````python
[5]*10 # Repetir un solo valor en una lista
````

````python
np.repeat( [1,2,3], 10  ) # repetir cada valor 10 veces
````

# Ciclos

Los ciclos son una forma de repetir un bloque de código un número determinado de veces o hasta que se cumpla una condición.


## For

Con los ciclos for, se debe especificar un arreglo en específico donde se realizará un recorrido por dicho arreglo.


````python
for i in range(5): # Recorrido desde 0 hasta 4
  print(i )
````

````python
for i in [1,3,5,7]: # Recorrido en la lista
  print( i)
````

````python
for i in np.array([ 0,1,7] ):
  print(i)
````

````python
for i in ["Pedro", "Antonio", "Sara" ]:
  print(i )
````

Cuando tenemos dos listas del mismo tamaño y queremos hacer un recorrido por ambas listas, range(len(lista)) nos permite tener una secuencia de todos los índices de la lista y con esta, se podrá realizar el recorrido por ambas listas.

````python
lista= ["Pedro", 15, "Juan", 20]
for i in range(len(lista) ):
  print( lista[i ])
````

````python
lista_1 = [20,30,50] # Lista 1
lista_2 = [500,60, 60] # lista 2
sum_list = [] # Lista vacia
for i in range(len(lista_1)): # Recorrido  desde 0 hasta la el tamaño de la lista -1
# código ejecución
  suma= lista_1[i] + lista_2[i] # sumamos los elementos
  sum_list.append(suma )  # Añadimos la suma a la nueva lista.
sum_list
````

## While

Este tipo de ciclos, realiza una secuencia de códigos mientras se cumpla la condición, en el momento que no se cumpla la condición termina el ciclo.

````python
x = 10
while x > 0 :  # Este ciclo se repite mientras el valor de x sea mayor que 0
    print(x) # Muestra resultados
    x = x - 1 # Código ejecución
````

# Condicionales if, else

Los condicionales if, elif y else, realizar la ejecución de un código si se cumple.


````python
x= 15
if x > 10:
  print("Es mayor que 10")
elif x<= 10:
  print( "Es menor o igual que 10")
else:
  print("Error")
````

````python
x= 5
if x > 10:
  print("Es mayor que 10")
elif x<= 10:
  print( "Es menor o igual que 10")
````

¿Cómo se pueden convinar los ciclos y condicionales?


Supongamos que se una lista de estudiantes, se tiene la nota de un examen final, puntos extra y se quiere saber la lista de los estudiantes que pasaron el examen sin los puntos extra y los que pasaron con los puntos extra.  

````python
lista_estudiantes= ["Alejandro", "Mateo","Luis", "Laura", "Tatiana", "Pedro", "Santiago","Luciana"] # Estudiantes
lista_notas= [4.2,3.2,2.5,3.4,2,1.5,2.9,2.9] # Nota del examen
lista_puntos_extra = [ 0 , 0.1,0.6,0.3,0.4,0.3,0,0.2] # Puntos adicionales
# Creamos la lista vacia
lista_aprobados_sin_puntos_extra=[]
lista_aprobados_con_puntos_extra=[]
for i in range(len(lista_estudiantes)):  # Paramos por un recorrido desde el primer estudiante hasta el último
  # Bloque de código
  nota =lista_notas[i]  # Nota del estudiante
  puntos_extra= lista_puntos_extra[i] # puntos extra
  if nota >=3 : # si la nota del examen mayor o igual a 3, aprueba sin puntos extra
    lista_aprobados_sin_puntos_extra.append(lista_estudiantes[i] )
    # sino se cumple entonces:
  elif(nota + puntos_extra ) >=3: # Si la nota mas los puntos extra son mayor o igual a 3, aprueba con puntos extra.
    lista_aprobados_con_puntos_extra.append(lista_estudiantes[i]  )

print("Lista de estudiantes que aprobaron sin puntos extra" )
print(lista_aprobados_sin_puntos_extra)
print("Lista de estudiantes que aprobaron con puntos extra")
print(lista_aprobados_con_puntos_extra )
````

# DataFrame

En esta sección se utilizaran las herramientas de pandas para crear columnas, filas, realizar filtros y eliminar registros.

Utilizando el ejemplo anterior ¿cómo es posible extraer dicha información si se tuviera la información en un Dataframe?

````python
datos_notas= pd.DataFrame({"notas":lista_notas,
                           "puntos extra": lista_puntos_extra},
                          index= lista_estudiantes  )
````

````python
datos_notas["notas"]+datos_notas["puntos extra"]  # Suma de columnas
````

````python
datos_notas.sum(axis=1 ) # Sumando todas las columnas
````

````python
datos_notas["nota final"] = datos_notas.sum(axis=1) # Crear una nueva columna
````

````python
filtro_aprobados_sin_puntos_extra = datos_notas["notas" ] >= 3
filtro_aprobados_con_puntos_extra =datos_notas["nota final"] >= 3
````

Una forma de filtrar el dataframe con pandas es utilizando [ ] o utilizando .loc[].


````python
datos_notas["nota final"] # seleccionar una columna
````

````python
filtro_aprobados_sin_puntos_extra # los estudiantes que aprobaron sin puntos extra
````

Cuando tenemos un dataframe con pandas, para seleccionar una columna utilizamos los corchetes.

```python
datos["Variable"]
```

Sin embargo nosotros podemos filtrar utilizando los chorchetes, donde se puede introducir un arreglo de booleanos.

```python
filtro= [ True, False, False, ..., True ]
datos[filtro]
```
Esto me arroja los datos con todas las columnas, pero aplicando el filtro y mostrando unicamente los individuos que tienen True en el filtro.



````python
datos_notas[filtro_aprobados_sin_puntos_extra] # Selecciona los que
````

````python
filtro_aprobados_con_puntos_extra # Los estudiantes que aprueban con puntos extra
````

````python
datos_notas[filtro_aprobados_con_puntos_extra] # No considera a Tatiana, Pedro y Santiago
````

Pero ¿cómo podemos hacer que seleccione los estudiantes que pasaron el examen gracias a los puntos extra?

````python
filtro_aprobados_por_puntos_Extra =  ~filtro_aprobados_sin_puntos_extra & filtro_aprobados_con_puntos_extra # A
````

````python
datos_notas[filtro_aprobados_por_puntos_Extra] # Filtrando los estudiantes que aprobaron gracias a los puntos extra
````

````python
datos_notas.loc[ filtro_aprobados_por_puntos_Extra, ["nota final"] ] # con .loc podemos realizar el filtro y seleccionar columnas
````
